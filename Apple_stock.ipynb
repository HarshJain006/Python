```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AAPL(80-24) Final.csv')
```

```python
df.head(5)
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
start_date ='12/31/2015'
end_date='12/31/2021'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
 
```

```python
filtered_df.set_index('Date', inplace=True)
monthly_avg = filtered_df['Low'].resample('M').mean().reset_index()

```

```python
sorted_df = filtered_df.sort_values(by='Date', ascending=True)
sorted_df
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Low',data=monthly_avg)
plt.title('Apple Stock Monthly Average Price (2010-2020)')
plt.xlabel('Date')
plt.ylabel('Average Close Price')
plt.xticks(rotation=45)
plt.grid()

plt.show()
```
