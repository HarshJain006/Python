Assignment 1: Define a class named Book that represents a book in a library.
Attributes :
title (string): The title of the book.
author (string): The author of the book.
year (integer): The year the book was published.
available (boolean): A status indicating whether the book is available or checked out.
Methods:
display_details() that prints the book’s details (title, author, and year).
borrow_book() that sets the availability to False and prints a message saying the book has been borrowed.
return_book() that sets the availability to True and prints a message saying the book has been returned.
Tasks:
Create at least two objects of the Book class with different attributes
Display the details of each book using the display_details method.
Use the borrow_book method on one of the objects and check its availability status.
Use the return_book method to change the status back to available.


```python
class Book:
    # Method to display book details
    def display_details(self):
        availability_status = "Available" if self.available else "Checked Out"
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Year: {self.year}")
        print(f"Status: {availability_status}\n")

    # Method to borrow a book (change availability to False)
    def borrow_book(self):
        if self.available:
            self.available = False
            print(f"The book '{self.title}' has been borrowed.")
        else:
            print(f"Sorry, the book '{self.title}' is already checked out.")

    # Method to return a book (change availability to True)
    def return_book(self):
        if not self.available:
            self.available = True
            print(f"The book '{self.title}' has been returned and is now available.")
        else:
            print(f"The book '{self.title}' is already available.")

# Creating two objects of the Book class without a constructor
book1 = Book()
book1.title = "The Great Gatsby"
book1.author = "F. Scott Fitzgerald"
book1.year = 1925
book1.available = True  # Manually assigning values

book2 = Book()
book2.title = "1984"
book2.author = "George Orwell"
book2.year = 1949
book2.available = True  # Manually assigning values

# Displaying details of both books
print("Displaying details of book 1:")
book1.display_details()

print("Displaying details of book 2:")
book2.display_details()

# Borrowing book1 and checking its availability
print("Borrowing book 1:")
book1.borrow_book()
book1.display_details()  # Checking if it's now unavailable

# Trying to borrow book1 again
print("Trying to borrow book 1 again:")
book1.borrow_book()

# Returning book1 and checking its availability again
print("\nReturning book 1:")
book1.return_book()
book1.display_details()  # Checking if it's now available
```

```python
class Shape:
    PI = 3.14

    def __init__(self, area=0, perimeter=0, length=0, breath=0, radii=0):
        self.area = area
        self.perimeter = perimeter
        self.length = length
        self.breath = breath
        self.radii = radii


class Circle(Shape):
    def __init__(self, radii):
        super().__init__(radii=radii)

    def area_circle(self):
        self.area = 2 * Shape.PI * self.radii
        print(self.area)


class Rect(Shape):
    def __init__(self, length, breadth):
        super().__init__(length=length, breath=breadth)

    def area_rect(self):
        self.area = self.length * self.breath
        print(self.area)


# s1 = Shape()
c1 = Circle(7)
c1.area_circle()

rect1 = Rect(5, 7)
rect1.area_rect()

# Polymorphism is a key concept in object-oriented programming (OOP) that allows methods
# to do different things based on the object it is acting upon, even though they share the same name.

# Without Polymorphism: We rely on type checks and manual selection of methods.
# With Polymorphism: Each object knows how to handle its own behavior, and no type checks are required.
```

Smart Home Devices Management
 Base abstract class: SmartDevice
Class variables: _name (string), _status
Abstract methods: turn_on(), turn_off()
Non-abstract method : device_info()
Subclass SmartLight: Inherit from the SmartDevice class.
Implement turn_on() and turn_off() method to set _status to 'on' and return "Smart light is now on.".
 Add a method set_brightness(level) that accepts a brightness level (integer from 0 to 100) and returns a string, e.g., "Brightness set to 70%.".
Use the super() function to initialize the inherited protected instance variables in the constructor.


```python
from abc import ABC, abstractmethod

# Abstract class SmartDevice
class SmartDevice(ABC):
    def __init__(self, name):
        self._name = name  # Protected instance variable
        self._status = "off"  # Default status is off

    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass

    def device_info(self):
        return f"Device Name: {self._name}, Status: {self._status}"


# Subclass SmartLight inheriting from SmartDevice
class SmartLight(SmartDevice):
    def __init__(self, name):
        super().__init__(name)  # Initialize _name and _status using super()

    def turn_on(self):
        self._status = "on"
        return "Smart light is now on."

    def turn_off(self):
        self._status = "off"
        return "Smart light is now off."

    def set_brightness(self, level):
        if 0 <= level <= 100:
            return f"Brightness set to {level}%."
        else:
            return "Error: Brightness level must be between 0 and 100."


# Example usage of SmartLight class
if __name__ == "__main__":
    # Create a SmartLight instance
    light = SmartLight("Living Room Light")

    # Turn the light on and display device info
    print(light.turn_on())
    print(light.device_info())

    # Set brightness and display the result
    print(light.set_brightness(70))

    # Turn the light off and display device info
    print(light.turn_off())
    print(light.device_info())

# Abstraction in Object-Oriented Programming (OOP) refers to the concept of hiding the complex
# implementation details and showing only the essential features or functionalities
# to the user. It helps in managing complexity by allowing you to focus on what an object
# does rather than how it does it.
```

```python
from abc import ABC, abstractmethod

# Abstract class SmartDevice
class SmartDevice(ABC):
    def __init__(self, name):
        self._name = name  # Protected instance variable
        self._status = "off"  # Default status is off

    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass

    def device_info(self):
        return f"Device Name: {self.__class__.__name__}, Status: {self._status}"


# Subclass SmartLight inheriting from SmartDevice
class SmartLight(SmartDevice):
    def __init__(self, name):
        super().__init__(name)  # Initialize _name and _status using super()

    def turn_on(self):
        self._status = "on"
        return "Smart light is now on."

    def turn_off(self):
        self._status = "off"
        return "Smart light is now off."

    def set_brightness(self, level):
        if 0 <= level <= 100:
            return f"Brightness set to {level}%."
        else:
            return "Error: Brightness level must be between 0 and 100."


# Subclass SmartThermostat inheriting from SmartDevice
class SmartThermostat(SmartDevice):
    def __init__(self, name):
        super().__init__(name)  # Initialize _name and _status using super()

    def turn_on(self):
        self._status = "on"
        return "Smart thermostat is ON."

    def turn_off(self):
        self._status = "off"
        return "Smart thermostat is OFF."

    def set_temperature(self, temp):
        return f"Temperature set to {temp}°C."

# Example usage of SmartLight and SmartThermostat classes
if __name__ == "__main__":
    # Create a SmartLight instance
    light = SmartLight("Living Room Light")

    # Turn the light on and display device info
    print(light.turn_on())
    print(light.device_info())

    # Set brightness and display the result
    print(light.set_brightness(70))

    # Turn the light off and display device info
    print(light.turn_off())
    print(light.device_info())

    # Create a SmartThermostat instance
    thermostat = SmartThermostat("Bedroom Thermostat")

    # Turn the thermostat on and display device info
    print(thermostat.turn_on())
    print(thermostat.device_info())

    # Set temperature and display the result
    print(thermostat.set_temperature(22))

    # Turn the thermostat off and display device info
    print(thermostat.turn_off())
    print(thermostat.device_info())
```
